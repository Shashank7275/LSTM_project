```python
index_to_word ={}
for word, index in word_index.items():
  index_to_word[index] = word
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def predictor(model,tokenizer,text,max_len):
  text = text.lower()
  seq = tokenizer.texts_to_sequences([text])[0]
  pad_seq = pad_sequences([seq], maxlen=max_len, padding="pre")

  pred = model.predict(pad_seq, verbose = 0)
  pred_index = np.argmax(pred)
  return index_to_word[pred_index]
```

```python
seed_text = "who are you"
next_word = predictor(lstm_model, tokenizer,seed_text,max_len)
print(next_word)
```

```python
def generate_quote(model,tokenizer,max_len,seed,total_words):
  seed_text = seed # Initialize seed_text with the seed parameter
  for _ in range(total_words):
    next_word = predictor(model,tokenizer,seed_text,max_len)
    if next_word ==" ":
      break
    seed_text += " "+ next_word
  return seed_text
```

```python
seed = "are you a  "
generated_text = generate_quote(lstm_model, tokenizer, max_len, seed, 10)
print(generated_text)
```
